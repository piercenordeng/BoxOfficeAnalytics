```python
import matplotlib.pyplot as plt
import collections
import pandas as pd
import numpy as np
import statsmodels.api as sm
import datetime
import calendar
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
rating = pd.read_csv("C:/Users/Zach's/GenreRatingUpdated2.0.csv", encoding="ISO-8859-1")
```

```python
rating.head()
```

```python
#Month
Studios = pd.get_dummies(rating['Studio'])
Genres = pd.get_dummies(rating['Genre'])
Directors = pd.get_dummies(rating['Director'])
MainActors = pd.get_dummies(rating['MainActor'])
#Ratings
```

```python
rating = pd.concat([rating, Studios], axis =1)
rating = pd.concat([rating, Genres], axis =1)
rating = pd.concat([rating, Directors], axis =1)
rating = pd.concat([rating, MainActors], axis =1)
```

```python
rating.head()
```

```python
y = rating[['TotalGross']]
x = rating.iloc[:,9:]

x = x.drop('Genre', axis = 1)

```

```python
x.head()
```

```python
X1 = sm.add_constant(x)

model = sm.OLS(y, X1)
result = model.fit()
predictions = result.predict()


model = sm.OLS(y, X1)

fitted = model.fit()
fitted.summary()
```
